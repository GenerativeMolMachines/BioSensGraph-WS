```python
import pandas as pd
import pubchempy as pcp
```

```python
df = pd.read_csv('../data/row/aptamers_from_aptagen.csv', index_col=0)
df_mol = df[df['Category'] == 'Small Organic']
```

```python
for idx, row in df_mol.iterrows():
    name = row['Target']
    compounds = pcp.get_compounds(name, namespace='name')

    for compound in compounds:
        # get IUPAC Name
        smiles = compound.isomeric_smiles
        df_mol.loc[idx, 'smiles'] = smiles
```

```python
filtered_df = df_mol[df_mol['smiles'].notnull()]
filtered_df.to_csv('../data/row/aptagen_small_molecules.csv')
```
